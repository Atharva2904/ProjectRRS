```python
thumbnails = tengen_df[tengen_df['Thumbnail'].notna()]
```

```python
# os.makedirs('thumbnails', exist_ok=True)
import requests
save_path = 'C:/SEM5/tempWorkspace/apurvinho/resources/thumbnails'

# Iterate over rows with non-empty thumbnail URLs
for _, row in thumbnails.iterrows():
    thumbnail_url = row['Thumbnail']
    book_index = row['BookIndex']
    
    try:
        # Send request to fetch the image
        response = requests.get(thumbnail_url, stream=True)
        if response.status_code == 200:
            # Save image with book index as filename
            with open(f'{save_path}/{book_index}.jpg', 'wb') as file:
                file.write(response.content)
        else:
            print(f"Failed to retrieve image for book index {book_index}")
    except Exception as e:
        print(f"Error for book index {book_index}: {e}")
```

```python
from PIL import Image, ImageDraw, ImageFont
import os
import random
initials_directory = "C:/SEM5/tempWorkspace/apurvinho/public/"  

# Function to create an image with a letter
def create_initial_image(letter, filename):
    # Generate a random background color
    bg_color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
    
    # Create an image with the random background color
    img = Image.new('RGB', (100, 100), color=bg_color)
    d = ImageDraw.Draw(img)
    
    # Use a default font
    font = ImageFont.load_default()  # You can specify a path to a TTF font file if needed
    
    # Calculate text size and position using textbbox for the bounding box
    bbox = d.textbbox((0, 0), letter, font=font)  # Get the bounding box for the text
    text_width = bbox[2] - bbox[0]  # width = right - left
    text_height = bbox[3] - bbox[1]  # height = bottom - top

    # Calculate text position to center it
    text_x = (img.width - text_width) // 2
    text_y = (img.height - text_height) // 2
    
    # Draw the letter on the image with white color
    d.text((text_x, text_y), letter, fill='white', font=font)
    
    # Save the image
    img.save(os.path.join(initials_directory, filename))

# Generate images for letters A-Z
for letter in range(65, 91):  # ASCII values for A-Z are 65-90
    create_initial_image(chr(letter), f"{chr(letter)}.jpeg")  # Create image for each letter

print("Initial letter images for A-Z with random backgrounds and white text generated successfully.")
```
